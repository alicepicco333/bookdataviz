```python
#How many events in Italy in 2022
import csv
from rdflib import Graph

def rdf_to_json(rdf_file, json_file):
    g = Graph()
    g.parse(rdf_file, format='xml')
    rdf_dict = {}
    for subj, pred, obj in g:
        subject = str(subj)
        predicate = str(pred)
        obj = str(obj)
        if subject not in rdf_dict:
            rdf_dict[subject] = {}
        rdf_dict[subject][predicate] = obj
    with open(json_file, 'w') as f:
        json.dump(rdf_dict, f, indent=4)

def json_to_dataframe(json_file):
    with open(json_file, 'r') as f:
        json_data = json.load(f)
    df = pd.DataFrame.from_dict(json_data, orient='index')
    return df

rdf_file = '../Information_visualization/dataset-eventi.rdf'
json_file = '../Information_visualization/dataset-eventi.json'

rdf_to_json(rdf_file, json_file)
df = json_to_dataframe(json_file)


# Load JSON data from file
with open('../Information_visualization/dataset-eventi.json', 'r') as file:
    data = json.load(file)

name_list = []

for item in data.values():
    if 'https://w3id.org/italia/onto/l0/name' in item:
        name_list.append(item['https://w3id.org/italia/onto/l0/name'])

file_path = 'name_list.csv'

# Writing the list to a CSV file
with open(file_path, mode='w', newline='', encoding='utf-8', errors='ignore') as file:
    writer = csv.writer(file)
    writer.writerows(name_list)


counter = 0
for item in name_list:
    counter += 1
print(counter)
```

```python
#How many of these were free

money_counter = 0
for item in data.values():
    if "https://w3id.org/italia/onto/POT/hasCurrencyValue" in item:
        if item["https://w3id.org/italia/onto/POT/hasCurrencyValue"] == "Gratuito":
            money_counter += 1
print(money_counter)
```

```python
#How many events per city 

city_dict = {}
for item in data.values():
    if "https://w3id.org/italia/onto/CLV/hasCity" in item:
        city_uri = item["https://w3id.org/italia/onto/CLV/hasCity"]
        city_name = city_uri.split("/City/")[-1]  # Extract the city name
        if city_name not in city_dict:
            city_dict[city_name] = 1
        else:
            city_dict[city_name] += 1

sorted_cities = sorted(city_dict.items(), key=lambda x: x[1], reverse=True)

for city, count in sorted_cities:
    print(f"{city}: {count}")
with open('city_events.json', 'w') as file:
    json.dump(city_dict, file, indent=4)
```

```python
from rdflib import Graph, Namespace
from rdflib.plugins.sparql import prepareQuery
from collections import defaultdict

# Create a graph
g = Graph()

# Load RDF data
g.parse("../Information_visualization/new-events.rdf")

# Define the namespace used in the RDF file
clvapit = Namespace("https://w3id.org/italia/onto/CLV/")

# Initialize defaultdict to store city counts
city_counts = defaultdict(int)

# Prepare SPARQL query
query = prepareQuery("""
    PREFIX clvapit: <https://w3id.org/italia/onto/CLV/>

    SELECT ?cityId (COUNT(?s) AS ?count)
    WHERE {
      ?s clvapit:hasCity ?city .
      BIND(STRAFTER(STR(?city), "/City/") AS ?cityId)
    }
    GROUP BY ?cityId
""")

# Execute SPARQL query
results = g.query(query)

# Iterate over query results and store city counts
for row in results:
    city_id = row["cityId"]
    count = row["count"]
    city_counts[city_id] = count

sorted_city_counts = dict(sorted(city_counts.items(), key=lambda item: item[1], reverse=True))

# Print city counts in descending order
for city_id, count in sorted_city_counts.items():
    print(f"Events in {city_id}: {count}")

with open('city_events.json', 'w') as file:
    json.dump(sorted_city_counts, file, indent=4)
```
